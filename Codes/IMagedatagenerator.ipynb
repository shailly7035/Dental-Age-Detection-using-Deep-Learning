```python
%matplotlib inline
import tensorflow as tf
import numpy as np
import cv2
import matplotlib.pyplot as plt
import pandas as pd
import os
from sklearn.utils import shuffle
from sklearn.model_selection import train_test_split
from keras.preprocessing.image import ImageDataGenerator , img_to_array
import pickle
import random
```

```python
NCLASSES = 9
HEIGHT = 300
WIDTH = 200
NUM_CHANNELS = 3
BATCH_SIZE = 10
```

```python
DATADIR = 'full frames/'
```

```python
def create_training_data():
    all_labels = pd.read_csv('all_output.csv')
    training_data = []
    for i in range(0,len(all_labels)):
        filename = str(all_labels.loc[i][0]) + '.jpg'
        label = all_labels.loc[i][10]
        #print(filename,label)
        training_data.append([filename,label])
    return training_data
```

```python
def main():
    train_datagen = ImageDataGenerator(rotation_range = 10,width_shift_range=0.1,zoom_range=0.1,horizontal_flip=True, fill_mode='nearest')
    training_data = create_training_data()
    training_data = random.sample(training_data , len(training_data))
    training_set , validation_set = train_test_split(training_data ,  random_state = 0, test_size = 0.2)
        #training_set = random.sample(training_set,130)
    train_X = []
    train_Y = []
    test_X = []
    test_Y = []
    for features,label in training_set:
        train_X.append(features)
        train_Y.append(label)

    for features,label in validation_set:
        test_X.append(features)
        test_Y.append(label)
        
    print(len(train_Y),len(test_Y))
    pickle_out = open( 'test_X.pickle' , 'wb')
    pickle.dump(test_X , pickle_out)
    pickle_out.close()
        
    pickle_out = open( 'test_Y.pickle' , 'wb')
    pickle.dump(test_Y , pickle_out)
    pickle_out.close()
        
    pickle_out = open( 'train_X.pickle' , 'wb')
    pickle.dump(train_X , pickle_out)
    pickle_out.close()
        
    pickle_out = open( 'train_Y.pickle' , 'wb')
    pickle.dump(train_Y , pickle_out)
    pickle_out.close()
        
        #continue
    c = pd.DataFrame({'Image': train_X,'Class':train_Y})
    c =c.astype(str)
    path = "output/" 
    i = 0
    for batch in train_datagen.flow_from_dataframe(dataframe = c,
        directory = "full frames",
        x_col = "Image",
        y_col = "Class",
        target_size=(HEIGHT , WIDTH), save_to_dir= path ,save_format='jpeg', save_prefix='N_',
            batch_size=BATCH_SIZE,class_mode = None):
        i += 1
        if i > 20: # save 20 images
            break  # otherwise the generator would loop indefinitely
```
